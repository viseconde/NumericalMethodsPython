```python
k = 0
x_list = []
y_list = []

try:
    point = int(input('Insira o número de pontos para determinar o polinômio: '))

    while k <= point-1:
        x = float(input('Insira o valor de x: '))
        y = float(input('Insira o respectiva ordenada y: '))
        x_list.append(x)
        y_list.append(y)
        k += 1
        
    dot = pd.DataFrame({
        "Abcissas": x_list,
        'Ordenadas': y_list
    })

except (TypeError, Exception, ZeroDivisionError) as error:
    print(f'Ocorreu um erro inesperado. (Erro: {error})')
```

Plotando o gráfico dos pontos discretos

```python
print('Índices', dot)
plt.figure(figsize=(10, 6))
plt.scatter(x_list, y_list)
plt.title('Pontos discretos')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

___lagr()___ retorna o valor de x na curva de interpolação

```python
c = lagrange(x_list, y_list)
lagr = poly(x_list, c)

t = np.linspace(min(x_list), max(x_list), 10)
lt = lagr(t) 
```
